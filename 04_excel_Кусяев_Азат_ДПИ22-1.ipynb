8. Напишите функцию `validate()`, которая проверяет соответствие всех строк из листа `Отзывы` следующим правилам:
    * Рейтинг - это число от 0 до 5 включительно
    * Соответствующий рецепт имеется на листе `Рецепты`
    
В случае несоответствия этим правилам, выделите строку красным цветом

```python
#Создаю список всех рецептов, присутствующих в листе Рецепты
recipes_list = sheet_recipes.range('B2').expand('down').value
recipes_list
```

```python
def validate(recipe):
    rating = recipe[4]
    recipe_id = recipe[2]
    if 0 <= rating <= 5 and recipe_id in recipes_list:
        return True
    return False
```

```python
review_rows = sheet_reviews.range('A2').expand('down')
for row in review_rows:
    if not validate(row.expand('right').value):
        row.expand('right').color = (255, 0, 0)
```
